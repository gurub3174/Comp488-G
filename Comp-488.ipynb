# DATA CLEANING

```python
def remove_high_null_columns(df, threshold=0.5):
    """
    Removes columns from df that have more than threshold% NaN values.
    
    Parameters:
        df (pd.DataFrame): The original DataFrame.
        threshold (float): The maximum allowed proportion of nulls (default 0.5).
        
    Returns:
        pd.DataFrame: A new DataFrame with columns having <= threshold nulls.
    """
    # Calculate the proportion of null values for each column.
    null_ratio = df.isnull().mean()
    
    # Identify columns to drop (those with more than threshold NaN proportion)
    columns_to_drop = null_ratio[null_ratio > threshold].index.tolist()
    
    # Print the columns that are going to be removed
    print(f"Columns to be removed (more than {threshold*100}% NaN values):")
    for col in columns_to_drop:
        print(f" - {col} ({null_ratio[col]*100:.2f}% nulls)")
    
    # Create a new DataFrame with the remaining columns
    new_df = df.drop(columns=columns_to_drop)
    return new_df

# Assuming your DataFrame is named Zillow2
clean_Zillow3 = remove_high_null_columns(Zillow3, threshold=0.5)
clean_Zillow2 = remove_high_null_columns(Zillow2, threshold=0.5)
clean_Zillow1 = remove_high_null_columns(Zillow1, threshold=0.5)
clean_Zillow = remove_high_null_columns(Zillow, threshold=0.5)


# # To confirm the remaining columns
# print("\nRemaining columns in the DataFrame:")
# print(clean_Zillow.columns.tolist())
# print(clean_Zillow1.columns.tolist())
# print(clean_Zillow2.columns.tolist())
# print(clean_Zillow3.columns.tolist())
```

# DATA STANDARDIZATION 

```python
# relevant numerical features we might use
relevant_columns = [
    'bedrooms', 'bathrooms', 'price', 'yearBuilt', 'livingArea', 'lotSize',
    'lotAreaValue', 'livingAreaValue', 'zestimate', 'rentZestimate', 
    'taxAssessedValue', 'taxAssessedYear', 'propertyTaxRate', 'isShowcaseListing', 
    'tourViewCount', 'lastSoldPrice', 'zestimateMinus30', 'restimateMinus30', 
    'zestimateLowPercent', 'zestimateHighPercent', 'restimateLowPercent', 
    'restimateHighPercent', 'dateSold', 'isRentalListingOffMarket', 'daysOnZillow', 
    'isConfirmedClaimedByCurrentSignedInUser'
]
exclude_columns = []
```

```python
def kmeans_impute(df, relevant_columns, n_clusters=5, random_state=42):

    df_copy = df.copy()
    
    missing_masks = {col: df_copy[col].isnull() for col in relevant_columns}
    
    # convert datetime columns to numeric and booleans to float.
    for col in relevant_columns:
        if 'date' in col.lower():
            df_copy[col] = pd.to_datetime(df_copy[col], errors='coerce')
            df_copy[col] = df_copy[col].apply(lambda x: x.timestamp() if pd.notnull(x) else np.nan)
            df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
        elif pd.api.types.is_datetime64_any_dtype(df_copy[col]):
            df_copy[col] = df_copy[col].apply(lambda x: x.timestamp() if pd.notnull(x) else np.nan)
            df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce')
        elif pd.api.types.is_bool_dtype(df_copy[col]):
            df_copy[col] = df_copy[col].astype(float)
    
    for col in relevant_columns:
        median_val = df_copy[col].median()
        df_copy[col] = df_copy[col].fillna(median_val)
    
    # Run KMeans clustering on the specified columns
    kmeans_model = KMeans(n_clusters=n_clusters, random_state=random_state)
    cluster_assignments = kmeans_model.fit_predict(df_copy[relevant_columns])
    df_copy['cluster'] = cluster_assignments
    
    # For each column and for each cluster, impute
    for col in relevant_columns:
        for cluster in range(n_clusters):
            cluster_mask = (df_copy['cluster'] == cluster)
            cluster_mean = df_copy.loc[cluster_mask, col].mean()
            mask_to_impute = missing_masks[col] & cluster_mask
            df_copy.loc[mask_to_impute, col] = cluster_mean
    
    df_copy = df_copy.drop(columns=['cluster'])
    
    return df_copy


Cleaned_combined_zillow = kmeans_impute(combined_Zillow, relevant_columns, n_clusters=5, random_state=42)
```
